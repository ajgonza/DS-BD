```python
# print the shape of the png image in grayscale and print the image
redweight=0.21
greenweight=0.72
blueweight=0.07

color_weight=[redweight, greenweight, blueweight]
img_png_grey=img_png.copy()
img_png_grey=np.dot(img_png_grey[:,:,:3],color_weight)
print("new image shape",img_png_grey.shape)
plt.imshow(img_png_grey,cmap='gray')
```

## Task 13

```python
# perform gamma correction on the image and display it
gamma = 1/5
img_png_gamma=img_png.copy()
img_png_gamma=img_png_gamma[:,:,:3]**gamma
plt.imshow(img_png_gamma)
```

```python
# perform gamma correction on the image and display it
gamma = 5
img_png_gamma5=img_png.copy()
img_png_gamma5=img_png_gamma5[:,:,:3]**gamma
plt.imshow(img_png_gamma5)
gamma = 1
img_png_gamma1=img_png.copy()
img_png_gamma1=img_png_gamma1[:,:,:3]**gamma
```

```python
# OPTIONAL - print 3 gamma corrected plots using subplots 

fig, ax=plt.subplots(1,3)

fig.suptitle('Horizontally stacked subplots')

ax[0].set_title('gamma =0.2')
ax[0].set_yticks([])
ax[0].imshow(img_png_gamma)
ax[1].set_title('gamma =1')
ax[1].set_yticks([])
ax[1].imshow(img_png_gamma1)
ax[2].set_title('gamma =5')
ax[2].set_yticks([])
ax[2].imshow(img_png_gamma5)

```
